```python
# 62
class Solution(object):
    def uniquePaths(self, m, n):
        """
        :type m: int
        :type n: int
        :rtype: int
        """
        dp = [[1 for _ in range(n)] for _ in range(m)]
        for i in range(1, m):
            for j in range(1, n):
                dp[i][j] = dp[i - 1][j] + dp[i][j - 1]
        return dp[-1][-1]
```

```python
# 63
class Solution(object):
    def uniquePathsWithObstacles(self, obstacleGrid):
        """
        :type obstacleGrid: List[List[int]]
        :rtype: int
        """
        n = len(obstacleGrid[0])
        m = len(obstacleGrid)
        dp = [[1 for _ in range(n)] for _ in range(m)]

        dp[0][0] = 1 - obstacleGrid[0][0]

        for i in range(1, m):
            if obstacleGrid[i - 1][0] == 1 or obstacleGrid[i][0] == 1:
                dp[i][0] = 0
            else:
                dp[i][0] = dp[i - 1][0]
                
        for j in range(1, n):
            if obstacleGrid[0][j - 1] == 1 or obstacleGrid[0][j] == 1:
                dp[0][j] = 0
            else:
                dp[0][j] = dp[0][j - 1]
                
        for i in range(1, m):
            for j in range(1, n):
                if obstacleGrid[i][j] == 1:
                    dp[i][j] = 0
                else:
                    dp[i][j] = dp[i - 1][j] + dp[i][j - 1]
        return dp[-1][-1]
```

```python
# 279
class Solution(object):
    def numSquares(self, n):
        """
        :type n: int
        :rtype: int
        """
        dp = [0] * (n + 1)
        dp[0] = 0
        for i in range(1, n + 1):
            start = int(math.sqrt(i))
            dp[i] = dp[i - 1] + 1
            for j in range(0, start + 1):
                try_one = dp[i - j ** 2] + 1
                if try_one < dp[i]:
                    dp[i] = try_one
        return dp[n]
```

```python
# 375
class Solution(object):
    def getMoneyAmount(self, n):
        """
        :type n: int
        :rtype: int
        """
        dp = [[0] * (n+1) for _ in range(n+1)] 
        for lo in range(n, 0, -1):
            for hi in range(lo+1, n+1):
                dp[lo][hi] = min(x + max(dp[lo][x-1], dp[x+1][hi]) for x in range(lo, hi))
        return dp[1][n]
```

```python
# 312
class Solution(object):
    def maxCoins(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        # k is the ballon that we want to blow it at last
        length = len(nums)
        nums = [1] + nums + [1]
        dp = [[0] * (length+2) for _ in range(length + 2)]
        for l in range(1, length+1):
            for i in range(1, length - l + 2):
                j = i + l - 1
                for k in range(i, j + 1):
                    dp[i][j] = max(dp[i][j], dp[i][k-1] + nums[i-1]*nums[k]*nums[j+1] + dp[k+1][j])
        return dp[1][length]
```

```python
# 322
class Solution(object):
    def coinChange(self, coins, amount):
        """
        :type coins: List[int]
        :type amount: int
        :rtype: int
        """
        if len(coins) < 0:
            return -1
        if len(coins) == 1:
            return amount // coins[0] if amount % coins[0] == 0 else -1
        dp = [0] * (amount + 1)
        
        for i in range(1, amount + 1):
            min_ = float("inf")
            for coin in coins:
                if i - coin >= 0:
                    res = dp[i - coin]
                    if res == -1:
                        continue
                    if res < min_:
                        min_ = 1 + res
            dp[i] = min_ if min_ != float("inf") else -1
        return dp[amount]

#
class Solution(object):
    def coinChange(self, coins, amount):
        """
        :type coins: List[int]
        :type amount: int
        :rtype: int
        """
        if len(coins) < 0:
            return -1
        dp = [float("inf")] * (amount + 1)
        dp[0] = 0
        
        for i in range(1, amount + 1):
            #dp[i] = min(dp[i - coin] for coin in coins)
            for coin in coins:
                if i - coin >= 0:
                    dp[i] = min(dp[i], dp[i - coin] + 1)
        #print(dp)
        return dp[amount] if dp[amount] <= amount else -1
```

```python
# 256
class Solution(object):
    def minCost(self, costs):
        """
        :type costs: List[List[int]]
        :rtype: int
        """
        dpr = dpb = dpg = 0
        for r, b, g in costs:
            dpr, dpb, dpg = min(dpb, dpg) + r, min(dpr, dpg) + b, min(dpr, dpb) + g
        return min(dpr, dpb, dpg)
```

```python
# 64
class Solution(object):
    def minPathSum(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        m = len(grid)
        n = len(grid[0])
        
        dp = [[0] * n for _ in range(m)]
        dp[0][0] = grid[0][0]
        
        for i in range(1, m):
            dp[i][0] = dp[i-1][0] + grid[i][0]
        for j in range(1, n):
            dp[0][j] = dp[0][j-1] + grid[0][j]
            
        for i in range(1, m):
            for j in range(1, n):
                dp[i][j] = min(dp[i-1][j], dp[i][j-1]) + grid[i][j]

        return dp[-1][-1]

# less space
class Solution(object):
    def minPathSum(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        m = len(grid)
        n = len(grid[0])
        
        
        for i in range(1, m):
            grid[i][0] += grid[i-1][0]
        for j in range(1, n):
            grid[0][j] += grid[0][j-1]
            
        for i in range(1, m):
            for j in range(1, n):
                grid[i][j] = min(grid[i-1][j], grid[i][j-1]) + grid[i][j]

        return grid[-1][-1]
```

```python
# 72
# recursive with memoery
class Solution(object):
    def minDistance(self, word1, word2):
        """
        :type word1: str
        :type word2: str
        :rtype: int
        """
        res = self.recur_helper(word1, word2, 0, 0, {})
        return res
        
    def recur_helper(self, word1, word2, i, j, memo):
        len1 = len(word1)
        len2 = len(word2)
        
        if i == len1 and j == len2:
            return 0
        elif i == len1:
            return len2 - j
        elif j == len2:
            return len1 - i
        
        if (i, j) not in memo:
            if word1[i] == word2[j]:
                ans = self.recur_helper(word1, word2, i + 1, j + 1, memo)
                memo[(i, j)] = ans
            else:
                insert = 1 + self.recur_helper(word1, word2, i, j+1, memo)
                delete = 1 + self.recur_helper(word1, word2, i + 1, j, memo)
                replace = 1 + self.recur_helper(word1, word2, i + 1, j + 1, memo)
                ans = min(insert, delete, replace)
                memo[(i, j)] = ans
        return memo[(i, j)]

# DP
class Solution(object):
    def minDistance(self, word1, word2):
        """
        :type word1: str
        :type word2: str
        :rtype: int
        """
        len1 = len(word1)
        len2 = len(word2)
        
        DP = [[0] * (len2 + 1) for _ in range(len1 + 1)]
        
        for i in range(1, len1 + 1):
            DP[i][0] = i
        for j in range(1, len2 + 1):
            DP[0][j] = j
            
        for i in range(1, len1 + 1):
            for j in range(1, len2 + 1):
                if word1[i - 1] == word2[j - 1]:
                    DP[i][j] = DP[i - 1][j - 1]
                else:
                    DP[i][j] = 1 + min(DP[i][j - 1], DP[i - 1][j], DP[i - 1][j - 1])
                    
        return DP[-1][-1]
```
