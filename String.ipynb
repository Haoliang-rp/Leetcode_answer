```python
# 68
# mine faster than 98%
class Solution(object):
    def fullJustify(self, words, maxWidth):
        """
        :type words: List[str]
        :type maxWidth: int
        :rtype: List[str]
        """
        res = []
        curr = []
        num = 0
        for word in words:
            len_word = len(word)
            if num + len_word > maxWidth:
                if curr:
                    res.append(self.helper(curr, maxWidth))
                    curr = [word]
                    num = len_word + 1
                else:
                    curr.append(word)
            
            else:
                curr.append(word)
                num += len_word + 1
        if curr:
            res.append(" ".join(curr))
            
        # deal with the last line
        last = res.pop()
        ans = last.strip().split()
        ans = " ".join(ans)
        res.append(ans + " " * (maxWidth - len(ans)))
        return res
        
    def helper(self, curr, maxWidth):
        """
        :type curr: List[str]
        :rtype: str
        """
        num = 0
        for word in curr:
            num += len(word)
        num_words = len(curr)
        if num_words == 1:
            return curr[0] + " " * (maxWidth - num)
        ans = ""
        num_space, remainder = divmod(maxWidth - num, num_words - 1)
        for i in range(num_words - 1):
            if remainder:
                ans += curr[i] + " " * (num_space + 1)
                remainder -= 1
            else:
                ans += curr[i] + " " * num_space
        return ans + curr[-1]
```

```python
# 76
class Solution(object):
    def minWindow(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: str
        """
        if not s or not t:
            return ""
        
        dict_t = collections.Counter(t)
        required = len(dict_t)
        
        l, r = 0, 0
        formed = 0
        window_counts = {}
        
        ans = (float("inf"), None, None)
        while r < len(s):
            character = s[r]
            window_counts[character] = window_counts.get(character, 0) + 1
            
            if character in dict_t and window_counts[character] == dict_t[character]:
                formed += 1
            
            while l <= r and formed == required:
                character = s[l]
                
                if r - l + 1 < ans[0]:
                    ans = (r - l + 1, l, r)
                
                window_counts[character] -= 1
                if character in dict_t and window_counts[character] < dict_t[character]:
                    formed -= 1
                    
                l += 1
            
            r += 1
        
        return "" if ans[0] == float("inf") else s[ans[1] : ans[2] + 1]
```

```python
# 30
class Solution(object):
    def findSubstring(self, s, words):
        """
        :type s: str
        :type words: List[str]
        :rtype: List[int]
        """
        if not words or not s:
            return []
        len_s = len(s)
        len_word = len(words[0])
        num_words = len(words)
        words_dict = collections.Counter(words)
        res = []
        
        # create several sliding window (of length len(words))
        for start in range(len_word):
            tmp_dict = words_dict.copy()
            matched = 0
            for end in range(start, len_s, len_word):
                cur_word = s[end : end + len_word]
                if cur_word in tmp_dict:
                    tmp_dict[cur_word] -= 1
                    matched += 1
                    while tmp_dict[cur_word] < 0:
                        pre_word = s[start : start + len_word]
                        tmp_dict[pre_word] += 1
                        matched -= 1
                        start = start + len_word
                    if matched == num_words:
                        res.append(start)
                else:
                    for k in range(start, end, len_word):
                        pre_word = s[k:k+len_word]
                        tmp_dict[pre_word] += 1
                    matched = 0
                    start = end + len_word
        return res
```

```python
# 3
class Solution(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        if not s:
            return 0
        check_dict = {}
        left, right = 0, 0
        ans = 0
        while right < len(s):
            if s[right] not in check_dict:
                check_dict[s[right]] = 1
                right += 1
                ans = max(ans, right - left)
            else:
                del check_dict[s[left]]
                left += 1
        return ans
```

```python
# 22
# backtracking
class Solution(object):
    def generateParenthesis(self, N):
        """
        :type n: int
        :rtype: List[str]
        """
        ans = []
        def backtrack(S = "", left = 0, right = 0):
            if len(S) == 2 * N:
                ans.append(S)
            if left < N:
                backtrack(S + "(", left + 1, right)
            if right < left:
                backtrack(S + ")", left, right + 1)
            
        backtrack()
        return ans

# DP
class Solution(object):
    def generateParenthesis(self, N):
        """
        :type n: int
        :rtype: List[str]
        """
        dp = [[] for _ in range(N + 1)]
        dp[0].append("")
        for i in range(N + 1):
            for j in range(i):
                dp[i] += ["(" + x + ")" + y for x in dp[j] for y in dp[i - j - 1]]
        return dp[N]
```

```python
# 241
class Solution(object):
    def diffWaysToCompute(self, input):
        """
        :type input: str
        :rtype: List[int]
        """
        if input.isdigit():
            return [int(input)]
        res = []
        for i in range(len(input)):
            if input[i] in "+-*":
                res1 = self.diffWaysToCompute(input[:i])
                res2 = self.diffWaysToCompute(input[i + 1:])
                for x in res1:
                    for y in res2:
                        res.append(self.oper(input[i], x, y))
        return res
        
            
    def oper(self, oper, one, two):
        if oper == "+":
            return one + two
        elif oper == "-":
            return one - two
        else:
            return one * two
```

```python
# 392
# two pointers
class Solution(object):
    def isSubsequence(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        if not s:
            return True
        if not t:
            return False
        i, j = 0, 0
        while i < len(s) and j < len(t):
            if s[i] == t[j]:
                i += 1
            j += 1
        return i == len(s)

# dict + bisect
class Solution(object):
    def isSubsequence(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        dict_t = collections.defaultdict(list)
        for i in range(len(t)):
            dict_t[t[i]].append(i)
            
        prev = 0
        for char in s:
            if char not in dict_t:
                return False
            else:
                idx_inT = dict_t[char]
                idx = bisect.bisect_left(idx_inT, prev)
                if idx >= len(idx_inT):
                    return False
                else:
                    prev = idx_inT[idx] + 1
        return True
```

```python
# 115
# https://www.youtube.com/watch?v=mPqqXh8XvWY
class Solution(object):
    def numDistinct(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: int
        """
        if not s:
            return 0
        if not t:
            return 1
        dp = [[None] * (len(s) + 1) for _ in range(len(t) + 1)]
        for i in range(1, len(t) + 1):
            dp[i][0] = 0
        for j in range(len(s) + 1):
            dp[0][j] = 1
        for i in range(1, len(t) + 1):
            for j in range(1, len(s) + 1):
                if t[i - 1] != s[j - 1]:
                    dp[i][j] = dp[i][j - 1]
                else:
                    # 1. don't use t[i - 1] == s[i - 1] 
                    # 2. use t[i - 1] == s[i - 1] 
                    dp[i][j] = dp[i][j - 1] + dp[i - 1][j - 1]                 
        return dp[-1][-1]
```
